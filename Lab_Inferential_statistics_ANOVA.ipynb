#### Context

Suppose you are working as an analyst in a microprocessor chip manufacturing plant. You have been given the task of analyzing a plasma etching process with respect to changing Power (in Watts) of the plasma beam. Data was collected and provided to you to conduct statistical analysis and check if changing the power of the plasma beam has any effect on the etching rate by the machine. You will conduct ANOVA and check if there is any difference in the mean etching rate for different levels of power. Y

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel('anova_lab_data.xlsx')
data
```

```python
data.columns = data.columns.str.replace(' ', '_').str.lower()
data
```

```python
data = data.rename(columns={'power_': 'power'})
```

```python
data['power'] = data['power'].str.replace(' W', '')
data
```

```python
data.groupby('power').agg(np.mean)
```

#### State the null hypothesis
#### State the alternate hypothesis
#### What is the significance level
#### What are the degrees of freedom of model, error terms, and to

<span style="color: blue;">**H0**:changing the power of the plasma beam **has No effect** on the etching rate by the machine.</span>.<hr>
<span style="color: blue;">**H1**:changing the power of the plasma beam **has an effect** on the etching rate by the machine.</span><hr>
<span style="color: blue;">Significance Level (alpha)= 0.05.</span> <hr>

<span style="color: blue;">**DOF**:3-1= 2</span>
<span style="color: blue;">**DOF Residual**: 15-3=12</span>

In this section, use Python to conduct ANOVA.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols


model = ols('etching_rate ~ C(power)',data=data).fit()
sm.stats.anova_lm(model)
```

The p-value is smaller than our alpha 0.05 so the H0 can be rejected. Changing the power does have an effect on the etching rate by the machine.

```python
data.pivot(columns='power').describe()
```

#### Let's see is power 200w is the reason that chaning the power has an effect on the etching rate

```python
from scipy.stats import ttest_ind

power_a = data[data['power'] == 200]['etching_rate']

for power in data['power'].unique():
    power_b = data[data['power'] == power]['etching_rate']
    print(power, ttest_ind(power_a, power_b))
    
```
